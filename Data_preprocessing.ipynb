```python
import pandas as pd
import numpy as np
import os 
```

```python
#data preprocessing for static scenario
data=pd.read_csv('D:/Research Working Folder/Paper_1-Mobile Sensing Virtual Environment/static_50cm.csv', header=3)

axisX=data[' Z [ m ]']
axisY=data['X [ m ]']
axisZ=data[' Y [ m ]']

data.drop(labels=['X [ m ]'], axis=1, inplace=True)
data.drop(labels=[' Y [ m ]'], axis=1, inplace=True)
data.drop(labels=[' Z [ m ]'], axis=1, inplace=True)

data.insert(0, 'X', axisX)
data.insert(1, 'Y', axisY)
data.insert(2, 'Z', axisZ)

data.rename(columns={' Temperature [ K ]':'0.0'}, inplace=True)

stepsize=1.0

for i in range(2400):
    a=data['0.0']
    a.name='{}'.format((i+1)*stepsize)
    data=pd.concat([data, a],axis=1)
```

```python
data.to_csv('D:/Research Working Folder/Paper_1-Mobile Sensing Virtual Environment/static_50cm_DEMO_new.csv', index=False)
```

```python
#data preprocessing for dynamic scenario
path='D:/Fluent/'

data=pd.read_csv(path+'FLTG-0001.1-8-00008-50cm')
axisX=data['    z-coordinate']
axisY=data['    x-coordinate']
axisZ=data['    y-coordinate']

data.drop(labels=['    x-coordinate'], axis=1, inplace=True)
data.drop(labels=['    y-coordinate'], axis=1, inplace=True)
data.drop(labels=['    z-coordinate'], axis=1, inplace=True)
data.drop(labels=['nodenumber'], axis=1, inplace=True)

data.insert(0, 'X', axisX)
data.insert(1, 'Y', axisY)
data.insert(2, 'Z', axisZ)

data.rename(columns={'     temperature':'0.0'}, inplace=True)
data['0.0']=300.0

stepsize=1.0
for i in range(1,2601,1):
    step_result=pd.read_csv(path+'FLTG-'+str('%04d' % i)+'.1-8-00008-50cm')
    a=step_result['     temperature']
    a.name='{}'.format((i)*stepsize)
    data=pd.concat([data, a],axis=1)
    
```
